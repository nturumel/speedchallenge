## Writing out individual frames 

```python
## Constants
CROP_IMG_DIR = os.path.join('cropped_images','train')
CROP_IMG_DIR_TEST = os.path.join('cropped_images','test')

TRAIN_VIDEO = os.path.join('data', 'train.mp4')
TEST_VIDEO = os.path.join("data","test.mp4")

TRAIN_VALUES = os.path.join('data', 'train.txt')
TRAIN_CSV = os.path.join("data","train.csv")
TEST_CSV = os.path.join("data","test.csv")

TRAIN_MD_CSV = os.path.join("data","train_md.csv")
TEST_MD_CSV = os.path.join("data","test_md.csv")
```

```python
# constants
train_frames = 20400
test_frames = 10798
```

```python
train_frame_names = []
tqdm.write('reading in training video file...')
tot_frames = train_frames
cap = cv2.VideoCapture(TRAIN_VIDEO)
idx = 0
tqdm.write('writing out frames...')
while True:
        ret, frame = cap.read()
        if not ret:
            break
        img_path = os.path.join(CROP_IMG_DIR, str(idx)+'.jpg')
        train_frame_names.append(img_path)
        skvideo.io.vwrite(img_path, frame)
        clear_output(wait=True)
        print("Processing frame number %d" % idx, end="\r")
        idx += 1
```

```python
test_frame_names = []
tqdm.write('reading in test video file...')
tot_frames = train_frames
cap = cv2.VideoCapture(TEST_VIDEO)
idx = 0
tqdm.write('writing out frames...')
while True:
        ret, frame = cap.read()
        if not ret:
            break
        img_path = os.path.join(CROP_IMG_DIR_TEST, str(idx)+'.jpg')
        test_frame_names.append(img_path)
        skvideo.io.vwrite(img_path, frame)
        clear_output(wait=True)
        print("Processing frame number %d" % idx, end="\r")
        idx += 1
```

## Dataset construction

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import csv
import skvideo.io
import json
from ast import literal_eval
from IPython.display import clear_output
from tqdm import tqdm
from sklearn.utils import shuffle
%matplotlib inline
```

```python
train_y = list(pd.read_csv(TRAIN_VALUES, header=None, squeeze=True))
assert(len(train_y)==train_frames)
```

```python
meta_dict = {}

for idx in range(0,train_frames - 1):
    meta_dict[idx] = [idx, train_frame_names[idx], train_frame_names[idx + 1], train_y[idx]]

train_meta_df = pd.DataFrame.from_dict(meta_dict, orient='index')
train_meta_df.columns = ['index', 'curr', 'next', 'speed']
train_meta_df.to_csv(TRAIN_MD_CSV, index=False)
```

```python
meta_dict = {}

for idx in range(0,test_frames - 1):
    meta_dict[idx] = [idx, test_frame_names[idx], test_frame_names[idx + 1]]

test_meta_df = pd.DataFrame.from_dict(meta_dict, orient='index')
test_meta_df.columns = ['index', 'curr', 'next']
test_meta_df.to_csv(TEST_MD_CSV, index=False)
```
