### Convert videos to individual frames

```python
## Constants
CROP_IMG_DIR = os.path.join('cropped_images','cropped_frame_')
CROP_IMG_DIR_TEST = os.path.join('cropped_images','cropped_test_frame_')
TRAIN_VIDEO = os.path.join('data', 'train.mp4')
TRAIN_VALUES = os.path.join('data', 'train.txt')
FRAME_TRAIN_CSV = os.path.join("data","frame_train.csv")
TRAIN_CSV = os.path.join("data","train.csv")
TEST_VIDEO = os.path.join("data","test.mp4")
FRAME_TEST_CSV = os.path.join("data","frame_test.csv")
TEST_CSV = os.path.join("data","test.csv")
CROP_RANGE = [215, 400, 30, 600]
RESIZE = [220, 66, 3]
```

```python
def generate_csv(data, filename):
    with open(filename, "w+") as f:
        writer = csv.writer(f)
        writer.writerows(data)
        f.close()
```

```python
def cut_resize(image):
    image_cropped = image[CROP_RANGE[0] : CROP_RANGE[1], CROP_RANGE[2] : CROP_RANGE[3]] #v1 for data
    image = cv2.resize(image_cropped, (RESIZE[0], RESIZE[1]), interpolation = cv2.INTER_AREA)    
    return image
```

```python
def split_train():
    speed_values = []
    with open(TRAIN_VALUES, "r") as f:
        for line in f:
            speed_values.append(literal_eval(line))
        
        
    video_reader = cv2.VideoCapture(TRAIN_VIDEO)
    ret, prev_img = video_reader.read()
    i = 0
    raw_data = []

    while True:
        ret, next_img = video_reader.read()
        i += 1
        if not ret:
            break

        # crop to frame
        crop_img_prev = cut_resize(prev_img)
        crop_img_next = cut_resize(next_img)
           
            
        filename_next = CROP_IMG_DIR + str(i) + r".png"
        filename_prev = CROP_IMG_DIR + str(i - 1) + r".png"
        
        cv2.imwrite(filename_next, crop_img_next)
        cv2.imwrite(filename_prev, crop_img_prev)
        
        prev_img = next_img
        
        raw_data.append([i, filename_next, filename_prev, speed_values[i]])
        
        clear_output(wait=True)
        print("Processing frame number %d\r" % i, end="\r")

    generate_csv(raw_data, FRAME_TRAIN_CSV)

split_train()
```

```python
def split_test():
    video_reader = cv2.VideoCapture(TEST_VIDEO)
    ret, prev = video_reader.read()
    i = 0
    raw_data = []

    while True:
        ret, next = video_reader.read()
        i += 1
        if not ret:
            break

        # crop to frame
        crop_img_prev = cut_resize(prev)
        crop_img_next = cut_resize(next)
           
            
        filename_next = CROP_IMG_DIR_TEST + str(i) + r".png"
        filename_prev = CROP_IMG_DIR_TEST + str(i - 1) + r".png"
        
        cv2.imwrite(filename_next, crop_img_next)
        cv2.imwrite(filename_prev, crop_img_prev)
        
        raw_data.append([i, filename_next, filename_prev])
        
        clear_output(wait=True)
        print("Processing frame number %d\r" % i, end="\r")

    generate_csv(raw_data, FRAME_TEST_CSV)

split_test()
```

### Preprocesssing

```python
# constants
INPUT_SHAPE = [RESIZE[1], RESIZE[0], RESIZE[2]]
CROP_IMG_OPFLOW = os.path.join('cropped_images','cropped_opflow_')
```

```python
def change_brightness(image, bright_factor = None):
    """
    Augments the brightness of the image by multiplying the saturation by a uniform random variable
    Input: image (RGB)
    returns: image with brightness augmentation
    """
    if not bright_factor:
        return image
    hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    # perform brightness augmentation only on the second channel
    hsv_image[:,:,2] = hsv_image[:,:,2] * bright_factor
    
    # change back to RGB
    image_rgb = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)
    return image_rgb
```

```python
def opticalFlowDense(image_current, image_next, index):
    """
    input: image_current, image_next (RGB images)
    calculates optical flow magnitude and angle and places it into HSV image
    * Set the saturation to the saturation value of image_next
    * Set the hue to the angles returned from computing the flow params
    * set the value to the magnitude returned from computing the flow params
    * Convert from HSV to RGB and return RGB image with same size as original image
    """
    gray_current = cv2.cvtColor(image_current, cv2.COLOR_RGB2GRAY)
    gray_next = cv2.cvtColor(image_next, cv2.COLOR_RGB2GRAY)
    
    
    hsv = np.zeros((66, 220, 3))
    # set saturation
    hsv[:,:,1] = cv2.cvtColor(image_next, cv2.COLOR_RGB2HSV)[:,:,1]
 
    # Flow Parameters
#     flow_mat = cv2.CV_32FC2
    flow_mat = None
    image_scale = 0.5
    nb_images = 1
    win_size = 15
    nb_iterations = 2
    deg_expansion = 5
    STD = 1.3
    extra = 0

    # obtain dense optical flow paramters
    flow = cv2.calcOpticalFlowFarneback(gray_current, gray_next,  
                                        flow_mat, 
                                        image_scale, 
                                        nb_images, 
                                        win_size, 
                                        nb_iterations, 
                                        deg_expansion, 
                                        STD, 
                                        0)
                                        
        
    # convert from cartesian to polar
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])  
        
    # hue corresponds to direction
    hsv[:,:,0] = ang * (180/ np.pi / 2)
    
    # value corresponds to magnitude
    hsv[:,:,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
    
    # convert HSV to float32's
    hsv = np.asarray(hsv, dtype= np.float32)
    rgb_flow = cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)
    
    filename = CROP_IMG_OPFLOW + str(index) + r".png"
    cv2.imwrite(filename, rgb_flow)


    
    return rgb_flow
```

```python
def load_image_from_path(image_path, bright_factor = None):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = change_brightness(img, bright_factor)    
    return img
```

```python
def generate_training_data(data, batch_size = 32):
    image_batch = np.zeros((batch_size, INPUT_SHAPE[0], INPUT_SHAPE[1], INPUT_SHAPE[2])) 
    label_batch = np.zeros((batch_size))
    while True:
        for i in range(batch_size):
            # generate a random index with a uniform random distribution from 1 to len - 1
            idx = np.random.randint(1, len(data) - 1)
            
            # Generate a random bright factor to apply to both images
            bright_factor = 0.2 + np.random.uniform()
            
            row = data.iloc[[idx]].reset_index()
            prev_img = load_image_from_path((row['Prev']).to_string(index=False))
            next_img = load_image_from_path((row['Next']).to_string(index=False))
            index = int((row['Index']).to_string(index=False))
            speed = float((row['Speed']).to_string(index=False))
            
           
            # compute optical flow send in images as RGB
            rgb_diff = opticalFlowDense(prev_img, next_img, index)
                        
            # calculate mean speed
            y = speed
            
            image_batch[i] = rgb_diff
            label_batch[i] = y
        
        # Shuffle the pairs before they get fed into the network
        yield shuffle(image_batch, label_batch)
```

```python
def generate_validation_data(data):
    while True:
        for idx in range(0, len(data) - 1):
            row = data.iloc[[idx]].reset_index()
            prev_img = load_image_from_path((row['Prev']).to_string(index=False))
            next_img = load_image_from_path((row['Next']).to_string(index=False))
            index = int((row['Index']).to_string(index=False))
            speed = float((row['Speed']).to_string(index=False))
            
           
            # compute optical flow send in images as RGB
            rgb_diff = opticalFlowDense(prev_img, next_img, index)
            rgb_diff = rgb_diff.reshape(1, rgb_diff.shape[0], rgb_diff.shape[1], rgb_diff.shape[2])
                        
            # calculate mean speed
            y =  np.array([[speed]])
       
            yield rgb_diff, y
```

```python
def get_model():
    model = Sequential()
    # normalization    
    # perform custom normalization before lambda layer in network
    model.add(Lambda(lambda x: x/ 127.5 - 1, input_shape = INPUT_SHAPE))

    model.add(Convolution2D(24, (5, 5), 
                            strides=(2,2), 
                            padding = 'valid',
                            kernel_initializer = 'he_normal',
                            name = 'conv1'))
    
    
    model.add(ELU())    
    model.add(Convolution2D(36, (5, 5), 
                            strides=(2,2), 
                            padding = 'valid',
                            kernel_initializer = 'he_normal',
                            name = 'conv2'))
    
    model.add(ELU())    
    model.add(Convolution2D(48, (5, 5), 
                            strides=(2,2), 
                            padding = 'valid',
                            kernel_initializer = 'he_normal',
                            name = 'conv3'))
    model.add(ELU())
    model.add(Dropout(0.5))
    model.add(Convolution2D(64, (3, 3), 
                            strides = (1,1), 
                            padding = 'valid',
                            kernel_initializer = 'he_normal',
                            name = 'conv4'))
    
    model.add(ELU())              
    model.add(Convolution2D(64, (3, 3), 
                            strides= (1,1), 
                            padding = 'valid',
                            kernel_initializer = 'he_normal',
                            name = 'conv5'))
              
              
    model.add(Flatten(name = 'flatten'))
    model.add(ELU())
    model.add(Dense(100, kernel_initializer = 'he_normal', name = 'fc1'))
    model.add(ELU())
    model.add(Dense(50, kernel_initializer = 'he_normal', name = 'fc2'))
    model.add(ELU())
    model.add(Dense(10, kernel_initializer = 'he_normal', name = 'fc3'))
    model.add(ELU())
    
    # do not put activation at the end because we want to exact output, not a class identifier
    model.add(Dense(1, name = 'output', kernel_initializer = 'he_normal'))
    
    adam = tf.optimizers.Adam(learning_rate=1e-4, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
    model.compile(optimizer = adam, loss = 'mse')

    print(model.summary())
    return model
```

```python
earlyStopping = EarlyStopping(monitor='val_loss', 
                              patience=3, 
                              verbose=1, 
                              min_delta = 0.23,
                              mode='min')


modelCheckpoint = ModelCheckpoint(WEIGHTS_LOC, 
                                  monitor = 'val_loss', 
                                  save_best_only = True, 
                                  mode = 'min', 
                                  verbose = 1,
                                 save_weights_only = True)

tensorboard = TensorBoard(log_dir=TENSORBOARD_LOC, histogram_freq=0,
                            write_graph=True, write_images=True)

callbacks_list = [modelCheckpoint, tensorboard, earlyStopping]
```

### Overfitting

```python
STEPS_PER_EPOCH = 100
EPOCHS = 8

train_data = train_data.sample(n=2500)
valid_data = valid_data.sample(n=500)

train_size = len(train_data.index)
train_generator = generate_training_data(train_data, BATCH_SIZE)
valid_generator = generate_validation_data(valid_data)
val_size = len(valid_data.index)

print('train size: ', train_size)
print('val size: ', val_size)

earlyStopping = EarlyStopping(monitor='loss', 
                              patience=3, 
                              verbose=1, 
                              min_delta = 0.23,
                              mode='min')


modelCheckpoint = ModelCheckpoint(WEIGHTS_LOC, 
                                  monitor = 'loss', 
                                  save_best_only = True, 
                                  mode = 'min', 
                                  verbose = 1,
                                 save_weights_only = True)

tensorboard = TensorBoard(log_dir=TENSORBOARD_LOC, histogram_freq=0,
                            write_graph=True, write_images=True)

callbacks_list = [modelCheckpoint, tensorboard, earlyStopping]

model = get_model()
history = model.fit(
        train_generator, 
        steps_per_epoch = STEPS_PER_EPOCH, 
        epochs = EPOCHS,
        callbacks = callbacks_list,
        verbose = 1)
```

```python
def make_predictions(data):
    predictions = []
    for idx in tqdm(range(0, len(data.index)-1)):
        row = data.iloc[[idx]].reset_index()
        prev_img = load_image_from_path((row['Prev']).to_string(index=False))
        next_img = load_image_from_path((row['Next']).to_string(index=False))
        index = int((row['Index']).to_string(index=False))
        speed = float((row['Speed']).to_string(index=False))
        
        
        img_diff = opticalFlowDense(prev_img, next_img, index)
        img_diff = img_diff.reshape(1, img_diff.shape[0], img_diff.shape[1], img_diff.shape[2])
        y = speed

        prediction = model.predict(img_diff)
        error = abs(prediction-y)
        
        
        #print(prediction, error, time_now)
        predictions.append([index, prediction, y, error])
        
    generate_csv(predictions, PREDCITIONS_CSV)
    predictions = pd.DataFrame(predictions)
    predictions.columns = ['Index', 'Prediction', 'Truth', 'Error']
    return predictions
```
