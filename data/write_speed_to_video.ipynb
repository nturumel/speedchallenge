```python
import pandas as pd

speed_data = pd.read_csv('results.csv')
speed_data.head()
```

```python
# Python program to write
# text on video
import cv2
from IPython.display import clear_output


idx = 0
cap = cv2.VideoCapture('test_half.mp4')
frames = []

while(True):	
    # Capture frames in the video
    ret, frame = cap.read()
   
    if idx == speed_data.shape[0]:
        break
    if not ret:
        break

    speed = str(speed_data.at[idx,'prediction'])
    speed = speed.replace("[","").replace("]","") 
    
    print("Processing frame number %d " % idx, end="\r")
    idx += 1
    
    
    # describe the type of font
    # to be used.
    font = cv2.FONT_HERSHEY_SIMPLEX
    # Use putText() method for
    # inserting text on video
    cv2.putText(frame,
                speed,
                (50, 50),
                font, 1,
                (0, 255, 255),
                2,
                cv2.LINE_4)

    frames.append(frame)
```

```python
print("Generating video")

# choose codec according to format needed
fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
video = cv2.VideoWriter('result.mp4', fourcc, 1, (640, 480))

for idx, frame in enumerate(frames):
    print("Processing frame number %d " % idx, end="\r")
    video.write(frame)


cv2.destroyAllWindows()
video.release()
```
